```python
nums = []
sims = {}
for i in range(1, 800):
    filename = f"../Data/test_cls/cls_zbin0_allrealizations_model{i}.npy"
    try:
        sims[i] = np.load(filename)
        nums.append(i)
    except Exception:
        pass
    
sims.keys()
```

```python
sims_mean = []
for i in nums:
    sims_mean.append(np.mean(sims[i], axis=0))
```

```python
params = np.load('../Data//params_cosmo_varied_om_s8_w0.npy')
params[1]
```

```python
params
```

```python
O_m = []
sigma8 = []
w0 = []

for i in nums:
    O_m.append(float(params[1][i - 1]))
    sigma8.append(float(params[2][i - 1]))
    w0.append(float(params[3][i - 1]))
```
